```python
cancer=load_breast_cancer()
```

```python
data=pd.DataFrame(cancer.data)
target=pd.Series(cancer.target)
```

```python
config=Config(30,16)
rtree=CART(target,data,config)
rtree.Fit("max")
```

```python
rtree
```

```python
pred=rtree.Predict(data)
```

```python
classy=[1 if i>0.5 else 0 for i in pred]
```

```python
plt.scatter(range(len(classy)),pred,c='r')
plt.show()
```

```python
plt.scatter(range(len(target)),target)
plt.show()
```

```python
row=[0 for i in range(len(pred))]
for i in range(len(target)):
    if classy[i]==target[i]:
        row[i]=1
    else:
        row[i]=0
```
