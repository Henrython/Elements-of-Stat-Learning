```python
from sklearn.tree import DecisionTreeRegressor
from CART import *
```

```python
from sklearn.datasets import load_boston
from matplotlib import pyplot as plt
```

```python
boston=load_boston()
```

```python
data=boston.data
```

```python
target=boston.target
```

```python
data=pd.DataFrame(data)
```

```python
data
```

```python
target=pd.Series(target)
```

```python
target
```

```python
data[3]=data[3].astype("int32").astype("category")
```

```python
data[8]=data[8].astype("int32").astype("category")
data=data[[1,2,4,5,6,7]]
```

```python
config=Config(10,1000)
rtree=CART(target,data,config)
rtree.Part()
```

```python
pred=rtree.Predict(data)
```

```python
pred
```

```python
plt.scatter(range(len(pred)),pred)
plt.show()
```

```python
plt.scatter(range(target.size),target)
plt.show()
```

```python
reg=DecisionTreeRegressor(max_depth=10)
```

```python
reg.fit(data,target)
```

```python
pred2=reg.predict(data)
```

```python
plt.scatter(range(len(pred2)),pred2)
plt.show()
```
